```python
# import necessary library to crawl data from steam
from bs4 import BeautifulSoup
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# avoid chromium crashing while crawl data on colab
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
def scrollDown(browser, numberOfScrollDowns):
    body = browser.find_element(By.TAG_NAME,'body')
    while numberOfScrollDowns >=0:
        body.send_keys(Keys.PAGE_DOWN)
        numberOfScrollDowns -= 1
        # print(numberOfScrollDowns)
    return browser
```

```python
data=[]
def crawl_url(url):
    browser=webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    browser.get(url)
    browser = scrollDown(browser, 100)
    html=browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    titles=soup.findAll('yt-formatted-string', class_='style-scope ytd-comment-renderer')   
    titles=[(link.text.replace('\t',' ')) for link in titles]
    titles=[(link.replace('\n',' ')) for link in titles]
    browser.quit()
    return titles 
```
